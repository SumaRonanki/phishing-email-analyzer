# EDA: LLM-Powered Phishing Email Analyzer
This notebook explores the dataset `data/raw/phishing_email.csv`.


```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# 1. Dataset Loading
df = pd.read_csv('../data/raw/phishing_email.csv')
display(df.head())
```

## Null Checks

```
# 2. Null Checks
print("Null values in dataset:")
print(df.isnull().sum())
```

## Class Distribution

```
# 3. Class Distribution Plot
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='label', palette='Set2')
plt.title('Class Distribution (0 = Legitimate, 1 = Phishing)')
plt.show()

print("Class Balance:")
print(df['label'].value_counts())
```

## Most Common Words

```
# 4. Most Common Words in Phishing vs Legitimate Emails
def get_most_common_words(text_series, num_words=20):
    all_text = ' '.join(text_series.astype(str).tolist())
    words = all_text.split()
    return Counter(words).most_common(num_words)

phishing_words = get_most_common_words(df[df['label'] == 1]['text_combined'])
legit_words = get_most_common_words(df[df['label'] == 0]['text_combined'])

print("Top 20 words in Phishing Emails:")
for word, count in phishing_words:
    print(f"{word}: {count}")
    
print("\nTop 20 words in Legitimate Emails:")
for word, count in legit_words:
    print(f"{word}: {count}")
```

## Email Length Comparison

```
# 5. Email Length Comparison
df['email_length'] = df['text_combined'].astype(str).apply(len)

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='label', y='email_length', palette='Set2')
plt.title('Email Length Comparison')
plt.ylim(0, 5000) # limiting y-axis for better visibility of the boxplot body
plt.show()

print("Average length of Phishing emails:", df[df['label'] == 1]['email_length'].mean())
print("Average length of Legitimate emails:", df[df['label'] == 0]['email_length'].mean())
```
